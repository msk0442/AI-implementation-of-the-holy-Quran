## 🧠 3. Advanced AI Solution Architecture (Autoencoders)

To solve this, we are going completely, using a specialized Neural Network called an **Autoencoder**.

**How does an Autoencoder work? (Explained simply!)**
Think of the Autoencoder as an artist who is forced to draw a highly detailed picture while looking through a tiny keyhole.
1. **The Encoder (Compressing):** The AI takes 10 different complex sensor readings (temperature, pressure, vibration) and mathematically squishes (compresses) them down into just 2 numbers! (This is the keyhole).
2. **The Bottleneck:** The data is now tiny. Only the most vital, essential 'patterns' of a healthy machine survive this compression.
3. **The Decoder (Rebuilding):** The AI then tries to rebuild the original 10 sensor readings perfectly using only those 2 tiny numbers.

**The Magic Trick for Anomaly Detection:**
We *only* train the AI on healthy, normal machine data. It becomes a key at rebuilding perfectly healthy data.
Suddenly, a part of the machine starts breaking (impending Takwir!). The data looks weird. The AI tries to compress it and rebuild it, but it fails miserably! The rebuilt data looks totally wrong.
We measure the difference between the original data and the rebuilt data (called the **Reconstruction Error**). If the error is massive, the AI sounds the alarm: *"DANGER! COLLAPSE IMMINENT!"*



## 🚀 Step 1: Environment Setup & Super-Dependencies

Let's load our AI toolbelt. We are using **PyTorch**, the absolute industry standard for Deep Learning, alongside data manipulation giants like Pandas and Numpy. Be prepared for greatness!



```python
# If running locally, you would run: !pip install torch numpy pandas matplotlib seaborn scikit-learn
import torch
import torch.nn as nn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Set our random seed so our results are reproducible (same every time we run it)
torch.manual_seed(42)
np.random.seed(42)

print("✅ Amazing! PyTorch and all Deep Learning visualizers are locked, loaded, and ready to prevent disasters.")

```

## 🎨 Step 2: Generating Realistic "Mock Data" (The Engine Sensors)

Because this must be a 100% self-contained open-source project, we cannot rely on you downloading a CSV file. We will write code to mathematically generate our own Jet Engine sensor data!

We will simulate **10 Sensors** (Vibration, Heat, RPM, Pressure, etc.).
1. We generate 5,000 minutes of **Healthy Data** (The engine running perfectly).
2. We simulate 100 minutes of **Impending Collapse Data** (Microscopic anomalies starting to happen right before the engine explodes).



```python
def generate_industrial_telemetry():
    print("⚙️ Initiating Simulation: Massive Industrial Turbine Telemetry...")

    # 1. Generate 5,000 rows of Healthy Data for 10 sensors
    # We use a normal distribution (bell curve) to simulate normal operating ranges
    healthy_data = np.random.normal(loc=50, scale=5, size=(5000, 10))

    # Let's add some complex interactions (e.g., if sensor 0 goes up, sensor 1 goes up)
    healthy_data[:, 1] = healthy_data[:, 0] * 1.5 + np.random.normal(0, 2, 5000)

    # 2. Generate 100 rows of Anomaly (Catastrophic Failure) Data
    # The anomaly might not be huge spikes, just slightly 'off' relationships
    anomaly_data = np.random.normal(loc=50, scale=5, size=(100, 10))
    # Break the relationship between sensor 0 and 1! (This represents a broken physical gear)
    anomaly_data[:, 1] = anomaly_data[:, 0] * 0.5 + np.random.normal(10, 5, 100)

    # Combine them (Healthy first, then Anomaly at the very end of the timeline)
    full_dataset = np.vstack([healthy_data, anomaly_data])

    # Create Labels: 0 for Healthy, 1 for Anomaly (For our testing later, the AI won't see this during training!)
    labels = np.array([0] * 5000 + [1] * 100)

    # Put it in a beautiful Pandas DataFrame
    column_names = [f"Sensor_{i+1}_Vibration/Heat" for i in range(10)]
    df = pd.DataFrame(full_dataset, columns=column_names)
    df['Ground_Truth_Failure'] = labels

    return df

engine_df = generate_industrial_telemetry()
print(f"✅ Simulation Complete! Total dataset size: {engine_df.shape[0]} timestamps.")
# Let's peek at the first 5 rows of our healthy engine!
display(engine_df.head())

```

## ⚙️ Step 3: Data Preprocessing (Scaling for the Neural Network)

Deep Learning models are very sensitive to large numbers. If Sensor 1 is reading '5000' and Sensor 2 is reading '0.5', the AI brain will ignore Sensor 2.
We use a `StandardScaler` to compress all sensor data so it ranges roughly between -2 and 2.

*Critical Concept:* We ONLY train the AI on Healthy data. We hide the anomalous data completely so the AI only knows what perfection looks like!



```python
# Separate our data
healthy_engine_data = engine_df[engine_df['Ground_Truth_Failure'] == 0].drop('Ground_Truth_Failure', axis=1).values
anomalous_engine_data = engine_df[engine_df['Ground_Truth_Failure'] == 1].drop('Ground_Truth_Failure', axis=1).values

# We scale the data based ONLY on the healthy data's parameters
scaler = StandardScaler()
healthy_scaled = scaler.fit_transform(healthy_engine_data)

# Apply that same scaling math to the anomaly data
anomalous_scaled = scaler.transform(anomalous_engine_data)

# Split healthy data into Training (to learn) and Validation (to test itself)
X_train, X_val = train_test_split(healthy_scaled, test_size=0.2, random_state=42)

# Convert Numpy arrays into PyTorch Tensors! (The magical data structure of Deep Learning)
tensor_X_train = torch.FloatTensor(X_train)
tensor_X_val = torch.FloatTensor(X_val)
tensor_anomalous = torch.FloatTensor(anomalous_scaled)

print("✅ Data scaled, split, and converted to Deep Learning Tensors. Ready to inject into the Neural Network!")

```

## 🧠 Step 4: Model Definition (Building the At-Takwir Neural Network)

Here is where the computer science happens. We write the Blueprint for our Autoencoder.
It has two parts:
* `Encoder`: Takes 10 inputs -> shrinks to 6 -> shrinks to a tiny bottleneck of **3**.
* `Decoder`: Takes the bottleneck of 3 -> expands to 6 -> aims to perfectly reconstruct the original **10**.



```python
class TakwirAutoencoder(nn.Module):
    def __init__(self, num_features):
        super(TakwirAutoencoder, self).__init__()

        # 1. THE ENCODER (Compressing reality into a tiny bottleneck)
        self.encoder = nn.Sequential(
            nn.Linear(num_features, 6),
            nn.ReLU(True), # Activation function allows the AI to learn complex curves, not just straight lines
            nn.Linear(6, 3) # The Absolute Bottleneck!
        )

        # 2. THE DECODER (Trying to rebuild reality from the bottleneck)
        self.decoder = nn.Sequential(
            nn.Linear(3, 6),
            nn.ReLU(True),
            nn.Linear(6, num_features) # Recreates the original 10 features
        )

    def forward(self, x):
        # Data flows from encoder to decoder
        encoded = self.encoder(x)
        decoded = self.decoder(encoded)
        return decoded

# Instantiate the AI Engine!
num_sensors = 10
ai_model = TakwirAutoencoder(num_features=num_sensors)

print(ai_model)
print("\n✅ Neural Network Architecture is completely assembled in PyTorch!")

```

## 🎓 Step 5: Training & Validation (Teaching the AI to understand "Healthy")

We will train the model for 50 epochs (loops).
* **The Loss Function:** Mean Squared Error (MSE). This measures the difference between the original data and the rebuilt data.
* **The Goal:** Make the MSE as close to 0 as possible for the healthy data!



```python
# We use MSE Loss to calculate the reconstruction error
criterion = nn.MSELoss()
# Adam Optimizer is the engine that updates the brain weights to make it smarter
optimizer = torch.optim.Adam(ai_model.parameters(), lr=0.01)

epochs = 50
train_losses = []
val_losses = []

print("🔥 Igniting Training Sequence...")

for epoch in range(epochs):
    # --- TRAINING PHASE ---
    ai_model.train()
    optimizer.zero_grad()

    # 1. AI tries to reconstruct the training data
    reconstructed_train = ai_model(tensor_X_train)
    # 2. Calculate how badly it messed up (The Error)
    loss = criterion(reconstructed_train, tensor_X_train)
    # 3. Mathematically figure out how to fix the brain (Backpropagation)
    loss.backward()
    # 4. Actually update the brain weights
    optimizer.step()

    train_losses.append(loss.item())

    # --- VALIDATION PHASE ---
    # Test it on healthy data it has never seen to ensure it is generalizing well
    ai_model.eval()
    with torch.no_grad():
        reconstructed_val = ai_model(tensor_X_val)
        val_loss = criterion(reconstructed_val, tensor_X_val)
        val_losses.append(val_loss.item())

    if (epoch+1) % 10 == 0:
        print(f"Epoch [{epoch+1}/{epochs}] | Train Error: {loss.item():.4f} | Validation Error: {val_loss.item():.4f}")

print("✅ Training Complete! The AI is now a master of detecting 'Healthy' vibrations.")

# Let's plot the learning curve! It should go down incredibly fast.
plt.figure(figsize=(10,4))
plt.plot(train_losses, label='Training Mistake Level', color='blue', linewidth=2)
plt.plot(val_losses, label='Validation Mistake Level', color='green', linestyle='dashed')
plt.title('AI Learning Curve: Mastering the Healthy State', fontsize=14, fontweight='bold')
plt.xlabel('Training Epochs')
plt.ylabel('Reconstruction Error (MSE)')
plt.legend()
plt.grid(True)
plt.show()

```

## 🗣️ Step 6: Testing Inference & Catching The Collapse!

This is where the magic happens! We are going to feed the AI completely continuous data. First, we feed it healthy data. Then, suddenly, we will feed it the **Anomaly Data**.

Because the AI has *never* seen Anomaly data, and the physical laws of the machine are breaking down (the "Folding Up" of At-Takwir), the AI will fail to reconstruct the data. The Error will spike massively, saving the day!



```python
# Prepare a continuous timeline dataset: 500 healthy readings followed by 100 anomaly readings
timeline_data_scaled = np.vstack([X_val[:500], anomalous_scaled])
tensor_timeline = torch.FloatTensor(timeline_data_scaled)

# Set model to evaluation mode
ai_model.eval()

# Run the entire timeline through the AI!
with torch.no_grad():
    timeline_reconstructed = ai_model(tensor_timeline)

# Calculate the Reconstruction Error for EVERY SINGLE timestamp mathematically
# Error = (Original - Rebuilt)^2
reconstruction_errors = torch.mean((tensor_timeline - timeline_reconstructed)**2, dim=1).numpy()

# ----------------- VISUALIZATION & ALERTING -----------------

# Set a threshold. If the error goes above this line, the machine is collapsing!
THRESHOLD = 1.0

plt.figure(figsize=(14, 6))

# Plot the errors
plt.plot(reconstruction_errors, color='black', alpha=0.7, label='Live AI Reconstruction Error')

# Paint the dots! Blue = Healthy, Red = Danger!
colors = ['blue' if error < THRESHOLD else 'red' for error in reconstruction_errors]
plt.scatter(range(len(reconstruction_errors)), reconstruction_errors, color=colors, s=20)

# Draw the Danger Threshold Line
plt.axhline(THRESHOLD, color='red', linestyle='dashed', linewidth=2, label='🔥 CATASTROPHIC TAKWIR THRESHOLD 🔥')

# Shade the area where the anomaly actually happens
plt.axvspan(500, 600, color='red', alpha=0.1, label='Actual System Collapse Phase')

plt.title('Project At-Takwir: Live Catastrophic Anomaly Detection System', fontsize=16, fontweight='bold')
plt.xlabel('Continuous Operational Time (Minutes)')
plt.ylabel('AI Confusion/Reconstruction Error')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()

print("\n🚨 AI AGENT REPORT:")
danger_count = np.sum(reconstruction_errors > THRESHOLD)
if danger_count > 0:
    print(f"CRITICAL WARNING! The AI detected {danger_count} moments of impending systemic collapse!")
    print("Action Taken: Autonomous shutdown initiated prior to hardware destruction. Millions of dollars saved.")
else:
    print("All systems nominal. Smooth sailing.")

```
