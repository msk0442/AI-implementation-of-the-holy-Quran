### 2. Architecture Selection: Building Our AI's Brain

Now, for the really exciting part: building the actual AI! This uses a powerful tool called **PyTorch** (think of it as a special toolkit for building AI brains) to create our `NabaSentinelNet`.

#### What is a Neural Network?

Imagine a neural network like a digital brain that learns from data. It has different layers of 'neurons' (like tiny processing units) that work together.

*   **`NabaSentinelNet` Class:** This is our custom AI brain. It's built with layers:
    *   **Input Layer:** This is where our four 'Quranic variables' (stability, sun, rain, temp gradient) go in. So, it has 4 'neurons'.
    *   **Hidden Layers:** These are the 'thinking' layers. Our AI has two of these, with 64 and 32 neurons. They use something called `ReLU` (Rectified Linear Unit) which is like a switch that turns on if the input is positive, helping the AI learn complex patterns. We also use `Dropout` (0.2) here, which is like briefly 'shutting off' some neurons during training. This prevents the AI from becoming too reliant on specific connections, making it smarter and more flexible.
    *   **Output Layer:** This is the final layer that gives us the answer. It has 1 neuron, and it uses `Sigmoid`. `Sigmoid` squishes the output into a number between 0 and 1, which is perfect for our 'Agricultural Success Score'.

#### Preparing the Data (Data Preprocessing)

Before the AI can learn, we need to get our data ready. Think of it like preparing ingredients before cooking:

*   **`X` and `y`:** We separate our data. `X` contains all the input features (stability, sun, rain, temp), and `y` is the 'answer' or target (agricultural success score).
*   **`train_test_split`:** We split our data into two groups: `training data` (what the AI learns from) and `testing data` (what we use to see how well it learned, like a final exam). We set aside 20% of the data for testing.
*   **`StandardScaler`:** This is like making all our ingredients the same size. For example, 'solar intensity' might be in the hundreds, while 'stability' is between 0 and 1. Scaling them makes sure no single input feature has too much influence just because its numbers are bigger.
*   **`torch.FloatTensor`:** PyTorch likes to work with its own special type of numbers called 'tensors'. So, we convert our prepared data into these tensors.

#### Training Our AI (Optimization & Training)

Now, it's time for the AI to learn!

*   **`model = NabaSentinelNet(...)`:** We create an instance of our AI brain.
*   **`criterion = nn.MSELoss()`:** This is our AI's 'report card'. `MSELoss` (Mean Squared Error Loss) tells the AI how wrong its predictions were. The smaller this number, the better the AI is doing.
*   **`optimizer = optim.Adam(...)`:** This is our AI's 'learning strategy'. The `Adam` optimizer helps the AI adjust its internal connections (weights) to get better at predicting, based on its 'report card' score.

*   **Training Loop (`epochs`):** This is like sending our AI to school for 100 days (`epochs`).
    *   In each 'day' (epoch), the AI makes predictions (`outputs = model(X_train_t)`).
    *   We check how wrong those predictions were using `loss = criterion(outputs, y_train_t)`.
    *   Then, the `loss.backward()` and `optimizer.step()` commands tell the AI to adjust its internal settings to make fewer mistakes next time. It's like correcting homework.

Every 20 epochs, we print the `Loss` to see how much smarter our AI is getting. You can see the loss number gets smaller and smaller, which means the AI is learning very well! Finally, we get a message that our AI model is trained – it has learned to recognize the 'balance of Earth's systems'!

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- THE DATA ENGINE (Creating the 'Mihād' Environment) ---
# We are generating a dataset that mimics the variables described in the Surah.

def generate_naba_intelligence_data(samples=2000):
    np.random.seed(42)

    # 1. Mountain Stability Index (Awtād) - Higher means more stable crust
    stability_index = np.random.uniform(0.1, 1.0, samples)

    # 2. Solar Radiance (Sīraj) - Solar intensity in Watts/m2
    solar_radiance = np.random.normal(800, 150, samples)

    # 3. Precipitation Intensity (Thajjājā) - Rainfall in mm
    precipitation = np.random.gamma(shape=2, scale=20, size=samples)

    # 4. Night-Day Thermal Gradient (Libāsā vs Ma'āshā)
    # The Surah mentions night as a 'covering' and day for 'livelihood'
    thermal_gradient = np.random.normal(15, 5, samples)

    # Target: Agricultural Viability (The 'Gardens' - Jannāt)
    # Logic: High stability + Moderate Sun + High Rain = High Viability
    yield_viability = (0.3 * stability_index) + (0.2 * (solar_radiance/1000)) + (0.5 * (precipitation/100)) + np.random.normal(0, 0.05, samples)
    yield_viability = np.clip(yield_viability, 0, 1)

    df = pd.DataFrame({
        'mountain_stability': stability_index,
        'solar_intensity': solar_radiance,
        'rainfall_mm': precipitation,
        'temp_gradient': thermal_gradient,
        'agricultural_success': yield_viability
    })

    return df

# Initialize Data
df = generate_naba_intelligence_data()
print("📊 Dataset Created! Displaying the 'Gardens of Abundance' variables:")
display(df.head())
```

```python
import torch

# --- PRODUCTION INFERENCE ENGINE ---
# This mimics how a real-world API would use our trained model.

def predict_crop_viability(stability, solar, rain, temp):
    """
    Takes environmental inputs and returns a viability score.
    """
    model.eval()
    # 1. Scaling the raw input (using the scaler we trained earlier)
    raw_input = np.array([[stability, solar, rain, temp]])
    scaled_input = scaler.transform(raw_input)
    input_tensor = torch.FloatTensor(scaled_input)

    # 2. Running Inference
    with torch.no_grad():
        prediction = model(input_tensor).item()

    return prediction

# Example Test: High Stability, Good Sun, Good Rain
sample_stability = 0.9  # Strong Mountain Roots
sample_solar = 850.0     # Radiant Sun
sample_rain = 60.0       # Abundant Clouds
sample_temp = 12.0       # Balanced Gradient

score = predict_crop_viability(sample_stability, sample_solar, sample_rain, sample_temp)

print("🚀 NABĀ-SENTINEL PRODUCTION INFERENCE REPORT")
print(f"Target Region Parameters: Stability={sample_stability}, Solar={sample_solar}, Rain={sample_rain}mm")
print(f"Calculated Viability Score: {score:.2%}")

if score > 0.7:
    print("Status: ✅ Highly Recommended for 'Gardens of Thick Foliage'")
else:
    print("Status: ⚠️ Region requires further environmental stabilization.")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# --- THE AI ARCHITECTURE (Neural Net for Global Stability) ---

class NabaSentinelNet(nn.Module):
    def __init__(self, input_dim):
      super(NabaSentinelNet, self).__init__()
      # Professional Deep Architecture
      self.layers = nn.Sequential(
          nn.Linear(input_dim, 64),
          nn.ReLU(),
          nn.Dropout(0.2),
          nn.Linear(64, 32), # Fix: Output from previous layer (64) should be input here
          nn.ReLU(),        # Adding another ReLU for the new hidden layer
          nn.Linear(32, 1),
          nn.Sigmoid() #Output as probability 0-1
      )

    def forward(self, x):
      return self.layers(x)

# Data Preprocessing
X = df.drop('agricultural_success', axis=1).values
Y = df['agricultural_success'].values.reshape(-1,1)


X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42) # Fix: use Y for y_train, y_test split
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test_t = scaler.transform(X_test)


# Convert to Tensors
X_train_t = torch.FloatTensor(X_train)
y_train_t = torch.FloatTensor(y_train)
X_test_t = torch.FloatTensor(X_test) # Fix: Convert X_test directly to tensor
y_test_t = torch.FloatTensor(y_test) # Fix: Convert y_test directly to tensor


# Initialize Model, Loss, and Optimizer
model = NabaSentinelNet(X_train.shape[1])
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)


# Training Loop
epochs = 100
for epoch in range(epochs): # Fix: Changed 'epch' back to 'epoch' for print statement consistency
  model.train()
  optimizer.zero_grad()
  outputs = model(X_train_t)
  loss = criterion(outputs, y_train_t)
  loss.backward()
  optimizer.step()

  if (epoch+1) % 20 == 0:
    print(f'Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}') # Fix: 'epochs' instead of 'epoch' in print

print("\n AI Model Trained to recognize the balance of Earth's systems!")
```

```python

def plot_results(y_test, predictions):
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=y_test.flatten(), y=predictions.flatten(), alpha=0.5, color='teal')
    plt.plot([0, 1], [0, 1], color='red', linestyle='--')
    plt.title("The Prophecy of Growth: Predicted vs Actual Agricultural Viability")
    plt.xlabel("Actual Environmental Health")
    plt.ylabel("AI Predicted Health")
    plt.grid(True)
    plt.show()

model.eval()
with torch.no_grad():
    preds = model(X_test_t).numpy()

plot_results(y_test, preds)
```
